```python
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('bill_survey.csv')
df.head()
```

```python
df.City.nunique()
```

```python
sns.histplot(df.Bill, kde = True)
```

```python
sns.histplot(df[df.City == 'Mumbai'], kde = True)
```

```python
sns.histplot(df[df.City == 'Surat'], kde = True)
```

```python
sample_means = []
for city , subdf in df.groupby('City'):
    sample_means.append(subdf.Bill.mean())

len(sample_means)
```

```python
sns.histplot(sample_means, kde = True)
```
